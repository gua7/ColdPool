```python
import numpy as np
import numpy.ma as ma
from matplotlib import pyplot as plt
```

```python
mean_the_r = np.loadtxt('new/mean_the01ref')
mean_the_p = np.loadtxt('new/mean_the01p4')
mean_the_m = np.loadtxt('new/mean_the01m4')
min_the_r = np.loadtxt('new/min_the01ref')
min_the_p = np.loadtxt('new/min_the01p4')
min_the_m = np.loadtxt('new/min_the01m4')
size_r = np.loadtxt('new/size_01ref')
size_p = np.loadtxt('new/size_01p4')
size_m = np.loadtxt('new/size_01m4')
mean_qt_r = np.loadtxt('new/mean_qt01ref')
mean_qt_p = np.loadtxt('new/mean_qt01p4')
mean_qt_m = np.loadtxt('new/mean_qt01m4')

t = np.arange(0,1800,1)/60
```

```python
min_the_r_masked = ma.masked_greater(min_the_r,-2)
min_the_p_masked = ma.masked_greater(min_the_p,-2)
min_the_m_masked = ma.masked_greater(min_the_m,-2)
```

```python
mean_the_r_masked = ma.filled(ma.masked_array(mean_the_r,mask = min_the_r_masked.mask), np.nan)
mean_the_p_masked = ma.filled(ma.masked_array(mean_the_p,mask = min_the_p_masked.mask), np.nan)
mean_the_m_masked = ma.filled(ma.masked_array(mean_the_m,mask = min_the_m_masked.mask), np.nan)
```

```python
mean_qt_r_masked = ma.filled(ma.masked_array(mean_qt_r,mask = min_the_r_masked.mask), np.nan)
mean_qt_p_masked = ma.filled(ma.masked_array(mean_qt_p,mask = min_the_p_masked.mask), np.nan)
mean_qt_m_masked = ma.filled(ma.masked_array(mean_qt_m,mask = min_the_m_masked.mask), np.nan)
median_mean_qt_r = np.nanmedian(mean_qt_r,axis = 1)
median_mean_qt_p = np.nanmedian(mean_qt_p,axis = 1)
median_mean_qt_m = np.nanmedian(mean_qt_m,axis = 1)
```

```python
numbers_r = np.zeros((len(min_the_r_masked[:,0])))
numbers_p = np.zeros((len(min_the_r_masked[:,0])))
numbers_m = np.zeros((len(min_the_r_masked[:,0])))

for i in range(len(min_the_r_masked[:,0])):
    numbers_r[i] = np.count_nonzero(~np.isnan(min_the_r_masked[i]))
    numbers_p[i] = np.count_nonzero(~np.isnan(min_the_p_masked[i]))
    numbers_m[i] = np.count_nonzero(~np.isnan(min_the_m_masked[i]))
```

```python
size_r=np.sqrt(size_r)
size_p=np.sqrt(size_p)
size_m=np.sqrt(size_m)
```

```python
size_r_masked = np.sqrt(ma.filled(ma.masked_array(size_r,mask = min_the_r_masked.mask), np.nan))
size_p_masked = np.sqrt(ma.filled(ma.masked_array(size_p,mask = min_the_p_masked.mask), np.nan))
size_m_masked = np.sqrt(ma.filled(ma.masked_array(size_m,mask = min_the_m_masked.mask), np.nan))
```

```python
median_size_r = np.nanmedian(size_r_masked,axis = 1)
median_size_p = np.nanmedian(size_p_masked,axis = 1)
median_size_m = np.nanmedian(size_m_masked,axis = 1)
```

```python
mean_size_r = np.nanmean(size_r_masked,axis = 1)
mean_size_p = np.nanmean(size_p_masked,axis = 1)
mean_size_m = np.nanmean(size_m_masked,axis = 1)
```

```python
max_size_r = np.nanmax(size_r_masked,axis = 1)
max_size_p = np.nanmax(size_p_masked,axis = 1)
max_size_m = np.nanmax(size_m_masked,axis = 1)
```

```python
n_size_r = np.nanpercentile(size_r_masked,90,axis = 1)
n_size_p = np.nanpercentile(size_p_masked,90,axis = 1)
n_size_m = np.nanpercentile(size_m_masked,90,axis = 1)
```
